```python
import os
import matplotlib.pyplot as plt

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd

audio_fpath = "../audio_data/"
audio_clips = os.listdir(audio_fpath)
print("No. of .wav files in audio folder = ",len(audio_clips))
```

```python
import pandas as pd

labels_df = pd.read_csv(audio_fpath+'esc50.csv')

labels_df.category.value_counts()
```

```python
x, sr = librosa.load(audio_fpath+audio_clips[2], sr=44100)

print(type(x), type(sr))
print(x.shape, sr)
```

#### Audio waveform in the time domain

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr)
```

#### Audio waveform to spectrogram

- visual representation of how the frequency content of an audio signal evolves over time
- one-dimensional time-domain audio waveform into a two-dimensional display for spacial characteristics

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

- X-axis: Represents time.
- Y-axis: Represents frequency.
- Intensity/Color:  amplitude (energy or loudness) of a particular frequency at a given time

#### Log transformation as scaling

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```
